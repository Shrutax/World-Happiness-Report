```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#for inline visualisation in Jupyter
%matplotlib inline
```

```python
import zipfile
```

```python
zip_path = "world-happiness-report.zip"
```

```python
with zipfile.ZipFile(zip_path, 'r') as z:
    print("Files in zip:")
    print(z.namelist())
    file_to_load = "2019.csv"
```

```python
#read specific file
with zipfile.ZipFile(zip_path, 'r') as z:
    with z.open(file_to_load) as f:
        df = pd.read_csv(f)
        
```

```python
#set the plot style
sns.set(style="whitegrid")
```

```python
#sort and select top 10
top10 = df.sort_values(by="Score", ascending = False).head(10)
```

```python
#Create bar plot
plt.figure(figsize=(12,6))
sns.barplot(x="Score", y="Country or region", data = top10, palette= "viridis")
```

```python
plt.title("Top 10 Happiest Countries - 2019" , fontsize=16)
plt.xlabel("happiness Score")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
#Correlation Heatmap(What Affects Happiness Most?)
import numpy as np
mask = np.triu(np.ones_like(df.corr(numeric_only=True), dtype=bool))
```

```python
#Heatmap
sns.heatmap(
    df.corr(numeric_only=True),
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",
    square=True,
    linewidths=0.9,
    cbar_kws={"shrink":0.15}
)
plt.title("Feature Correlation with Happiness Score", fontsize=18,pad=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
#We observe that higher GDP is generally associated with greater happiness.
```
